```python
def maxWidthNoMap(head):
    if not head:
        return 0
    maxWidth = 0
    curWidth = 0
    queue = deque()
    queue.append(head)
    cur_level = head
    next_level = None
    while queue:
        cur = queue.popleft()
        curWidth += 1
        if cur.left:
            next_level = cur.left
        if cur.right:
            next_level = cur.right
        if cur == cur_level:
            maxWidth = max(maxWidth, curWidth)
            curWidth = 0
            cur_level = next_level
    return maxWidth
    
```

```python
#题意给的Node类
class Node:
    def __init__(self, val, p=None, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
        self.parent = p
```

```python
def getLeftMost(node):
    while node.left:
        node = node.left
    return node

def getSuccessorNode(node):
    if not node:
        return None
    if node.right:
        return getLeftMost(node.right)
    else:
        parent = node.parent
        while parent and parent.right == node:
            node = parent
            parent = node.parent
        return parent
            
```
